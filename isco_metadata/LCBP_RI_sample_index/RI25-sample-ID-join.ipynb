### Sample IDs from Metadata/

        - A csv with the RI24 sample identifier and date/time/site/sample type

### Sample chemistry from LCBP-EMMA/
       
- LCBP_RI25_icpoes-for-join.csv
        - A csv with all RI25 inductively coupled plasma optical emission spectroscopy results Dartmouth SPECTRO
        
- LCBP_RI25_isotope-for-join.csv
        - A csv with all RI24 stable water isotope results from MN lab

```python
import os
os.chdir("/home/millieginty/OneDrive/git-repos/EMMA/data/newrnet-chemistry/RI25/")
```

```python
import pandas as pd

# Read CSV files with the 'Sample ID' column as the index
df_index = pd.read_csv('LCBP_RI25_sample-index-for-join.csv', index_col='Sample ID')
df_icpoes = pd.read_csv('ICPOES/LCBP_RI25-ICP-for-join.csv', index_col='Sample ID')
#df_ic = pd.read_csv('IC/LCBP_RI25_IC_Cl-for-join.csv', index_col='Sample ID')
df_iso = pd.read_csv('stable-isotopes/LCBP_RI25_isotope-for-join.csv', index_col='Sample ID')

# Join the dataframes on the index ('Sample ID')
df_joined = df_index.merge(df_icpoes, left_index=True, right_index=True, how='left')
#df_joined = df_joined.merge(df_ic, left_index=True, right_index=True, how='left')
df_joined = df_joined.merge(df_iso, left_index=True, right_index=True, how='left')

# Save the joined dataframe to a new CSV file
df_joined.to_csv('RI25-ICP-isotope-joined.csv')

# Print the first few rows of the joined dataframe
df_joined.head(50)
```
